### Context
The Philippine Statistics Authority (PSA) spearheads the conduct of the Family Income and Expenditure Survey (FIES) nationwide. The survey, which is undertaken every three (3) years, is aimed at providing data on family income and expenditure, including, among others, levels of consumption by item of expenditure, sources of income in cash, and related information affecting income and expenditure levels and patterns in the Philippines.


### Content
Inside this data set is some selected variables from the latest Family Income and Expenditure Survey (FIES) in the Philippines. It contains more than 40k observations and 60 variables which is primarily comprised of the household income and expenditures of that specific household

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

logistic_model = LogisticRegression(class_weight="auto")
logistic_model.fit(scaled_features_df, family_df['is_household_head_female'])

predictions = logistic_model.predict(scaled_features_df)
accuracy = metrics.accuracy_score(predictions, family_df['is_household_head_female'])

print(f'\nCoefficient values: {logistic_model.coef_}')
print(f'\nAccuracy = {accuracy * 100:.4f}')

features = len(logistic_model.coef_[0])
print(f'Total number of features: {features}\n')

coeffs = pd.DataFrame(list(zip(scaled_features_df.columns, logistic_model.coef_[0])))
print('\nSorted coefficient values:\n')
coeffs.sort_values(by=[1])
```

```python
# prepare selection list with the top 2 positively correlated beta coefficients, 
# bottom 2 negatively correlated beta coefficients, along with the Y and probability columns

selection_lst = []

beta_coeffs = list(coeffs[0])
    
for row in beta_coeffs:
    selection_lst.append(row)
    
selection_lst.append('is_household_head_female')
selection_lst.append('prob')

# selection_lst
```

```python
family_df['logistic_model_preds'] = predictions
prob =[x[1] for x in logistic_model.predict_proba(scaled_features_df)]

family_df['prob'] = prob

selection = family_df[selection_lst]

print('\nProbabilities of each Household Head being Female:\n')
selection[(selection['is_household_head_female'] == True)].sample(10).sort_values(by=['prob'], ascending=False)
```

```python
from sklearn.model_selection import train_test_split

logistic_model = LogisticRegression(class_weight='balanced')

X_train, X_test, Y_train, Y_test = train_test_split(scaled_features_df, family_df['is_household_head_female'], train_size=0.2)

logistic_model.fit(X_train, Y_train)

pred_train = logistic_model.predict(X_train)
pred_test =  logistic_model.predict(X_test)
train_accuracy = metrics.accuracy_score(pred_train, Y_train)
test_accuracy = metrics.accuracy_score(pred_test, Y_test)

print(f'\nTrain Accuracy = {train_accuracy * 100:.4f}')
print(f'\nTest Accuracy = {test_accuracy * 100:.4f}')
print(f'Test Precision Score = {metrics.precision_score(Y_test, pred_test) * 100:.4f}')
print(f'Test Recall Score = {metrics.recall_score(Y_test, pred_test) * 100:.4f}\n')

from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(Y_train, pred_train)
cnf_matrix

class_names=["0","1"] # name of classes
fig, ax = plt.subplots(figsize=(7, 6))
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

# labels = ['True Neg','False Pos','False Neg','True Pos']
# labels = np.asarray(labels).reshape(2,2)

# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('\nConfusion matrix\n')
plt.ylabel('\nActual label\n')
plt.xlabel('\nPredicted label\n')
```

## KMeans Clustering
---

```python
plt.figure(figsize=(20,10))
X = np.array(family_df[['Total Income from Entrepreneurial Acitivites', 'Total Household Income']].values.tolist())
```

```python
from sklearn.cluster import KMeans

var = 5

kmeans = KMeans(n_clusters=var)
kmeans.fit(X)

y_kmeans = kmeans.predict(X)

# init KMeans DF
km_df = pd.DataFrame()
km_df = family_df

# populate DF
km_df["cluster"] = y_kmeans
```

```python
km_df.head(5)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.xlabel("Mean Total Income from Entrepreneurial Acitivites\nper Total Household Income")
plt.ylabel("Total Household Income")
plt.title("\nRelationship between\nTotal Income from Entrepreneurial Acitivites\nand Total Household Income\n")
```

```python
palette = sns.color_palette("colorblind", var).as_hex()
palette
```

```python
color_getter = lambda x: palette[int(x)]
km_df['color'] = km_df['cluster'].map(color_getter)

km_df.head()
```

```python
sum_of_squared_distances = []
K = range(1,20)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    sum_of_squared_distances.append(km.inertia_)
```

```python
plt.figure(figsize=(10,8))
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.ylabel('Sum_of_squared_distances')
plt.title('\nElbow Method For Optimal k\n')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil = []
kmax = 20

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
  kmeans = KMeans(n_clusters = k).fit(X)
  labels = kmeans.labels_
  sil.append(silhouette_score(X, labels, metric = 'euclidean'))
```

### The Silhouette Score reaches its global maximum at the optimal k.
>The silhouette value measures how similar a point is to its own cluster (cohesion) compared to other clusters (separation).

### Reference:
- https://medium.com/analytics-vidhya/how-to-determine-the-optimal-k-for-k-means-708505d204eb
