## Data Cleaning 
---

```python
# check which cols have NULL entries

null_cols = family_df.columns[family_df.isna().any()].tolist()
print(f'\nColumns with NULL entries: {null_cols}\n')
```

```python
mask = family_df['Household Head Occupation'].isnull() & family_df['Household Head Class of Worker'].isnull()
family_df.loc[mask]
```

```python
# drop rows with null entries
family_df = family_df.dropna()

# drop rows with duplicates
family_df = family_df.drop_duplicates()
```

```python
target = 'Total Household Income'
k = 10

corr_vals = family_df.corr()
cols = corr_vals.nlargest(k, target)[target].index

corr_coeffs = np.corrcoef(family_df[cols].values.T)

ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(corr_coeffs, cmap='coolwarm', annot=True, fmt='.2f', yticklabels=cols.values, xticklabels=cols.values)
ax.set_title('\nTop 10 Variables Most Correlated\nwith Total Household Income\n', fontsize=30)
```

## Linear Regression
---

```python
from sklearn.linear_model import LinearRegression

# init DF with top 8 most correlated values with Total Household Income
X = family_df[cols]
X = X.drop(['Total Household Income (Monthly)', 'Total Household Income'], axis=1)
X.head(5)
```

```python
X.describe()
```

## Scale beta coefficients
---

```python
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import StandardScaler

mapper = DataFrameMapper([(X.columns, StandardScaler())])
scaled_features = mapper.fit_transform(X.copy())
scaled_features_df = pd.DataFrame(scaled_features, index=X.index, columns=X.columns)

scaled_features_df.describe()
```

```python
scaled_features_df.head(5)
```

```python
# determine the set of beta coefficients that can be multiplied to each 
# row to acquire the temperature value 
linear_model = LinearRegression()
linear_model.fit(scaled_features_df, family_df['Total Household Income'])
```

```python
linear_model.coef_
```

```python
# display how many estimated coefficient values were generated and 
# visualize them against the Dataset columns

print(f'\nNumber of estimated coefficient values: {len(linear_model.coef_)}\n')
print(f"Coefficient values: {linear_model.coef_}\n")

coeffs = pd.DataFrame(list(zip(X.columns, linear_model.coef_)))
coeffs.sort_values(by=[1])
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,8))

# plt.plot([0, 2000000], [0, 15000000], '--go')
sns.regplot(family_df['Total Income from Entrepreneurial Acitivites'], family_df['Total Household Income'])
plt.xlabel("Mean Total Income from Entrepreneurial Acitivites\nper Total Household Income")
plt.ylabel("Total Household Income")
plt.title("\nRelationship between\nTotal Income from Entrepreneurial Acitivites\nand Total Household Income\n")
```

## Predict Total Household Income using Beta Coefficients
---

```python
predictions = linear_model.predict(scaled_features_df)
family_df['PRED'] = predictions

family_df[['Total Household Income', 'PRED']].head()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,8))

sns.regplot(family_df['Total Household Income'], family_df['PRED'])
# plt.plot([0, 12000000], [0, 300000000000], '--go')
plt.xlabel("Actual Total Household Income")
plt.ylabel("Predicted Total Household Income")
plt.title("\nActual vs Predicted Total Household Income\n")
```

## KMeans Clustering
---

```python
plt.figure(figsize=(20,10))
X = np.array(family_df[['Total Income from Entrepreneurial Acitivites', 'Total Household Income']].values.tolist())
```

```python
from sklearn.cluster import KMeans

var = 10

kmeans = KMeans(n_clusters=var)
kmeans.fit(X)

y_kmeans = kmeans.predict(X)

# init KMeans DF
km_df = pd.DataFrame()
km_df = family_df

# populate DF
km_df["cluster"] = y_kmeans
```

```python
km_df.head(5)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.xlabel("Mean Total Income from Entrepreneurial Acitivites\nper Total Household Income")
plt.ylabel("Total Household Income")
plt.title("\nRelationship between\nTotal Income from Entrepreneurial Acitivites\nand Total Household Income\n")
```

```python
palette = sns.color_palette("colorblind", var).as_hex()
palette
```

```python
color_getter = lambda x: palette[int(x)]
km_df['color'] = km_df['cluster'].map(color_getter)

km_df.head()
```

```python
sum_of_squared_distances = []
K = range(1,20)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    sum_of_squared_distances.append(km.inertia_)
```

```python
plt.figure(figsize=(10,8))
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.ylabel('Sum_of_squared_distances')
plt.title('\nElbow Method For Optimal k\n')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil = []
kmax = 20

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
  kmeans = KMeans(n_clusters = k).fit(X)
  labels = kmeans.labels_
  sil.append(silhouette_score(X, labels, metric = 'euclidean'))
```

### The Silhouette Score reaches its global maximum at the optimal k.
>The silhouette value measures how similar a point is to its own cluster (cohesion) compared to other clusters (separation).

### Reference:
- https://medium.com/analytics-vidhya/how-to-determine-the-optimal-k-for-k-means-708505d204eb

```python
# K = np.array(K)
# sil = np.array(sil)

plt.figure(figsize=(10,8))
plt.plot(K, sil, 'mx-')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('\nSilhouette Method For Optimal k\n')
plt.show()
```
